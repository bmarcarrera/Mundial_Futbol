# <u>Estadísticas generales ganadores de los mundiales de fútbol</u>

A continuación, se presentan los datos más relevantes de todos los mundiales de fútbol celebrados, ordenados en función de la selección ganadora en cada evento, a partir de un *archivo.csv* que contiene la información más relevante de cada Mundial celebrado desde 1930 hasta 2022.

Se ha generado un DataFrame limpio utilizando la librería Pandas de Python utilizando los siguientes parámetros:

    - Campeón
    - Copas ganadas
    - Años ganados
    - Subcampeones
    - Máximo goleador del campeonato
    - País anfitrión

## Importar Pandas


```python
import pandas as pd
```

## Importar archivo *.csv*

```python
mundiales_csv = "https://github.com/bmarcarrera/Mundial_Futbol/blob/main/world_cup.csv"
```

## Crear *DataFrame* con *Pandas*

Se genera un *DataFrame* a partir del archivo importado y poder manipular los datos de forma más visual y sencilla.


```python
df_mundiales = pd.read_csv(mundiales_csv)
df_mundiales

```

## Cambiar el nombre de determinados valores

Se cambia al idioma español los nombres de los países de forma manual.

```python
paises = {
    "France": "Francia",
    "Croatia": "Croacia",
    "Netherlands":  "Holanda",
    "Germany": "Alemania",
    "West Germany": "Alemania",
    "Brazil": "Brasil",
    "Italy": "Italia",
    "Czechoslovakia": "Checoslovaquia",
    "Sweden": "Suecia",
    "Hungary": "Hungría",
    "Switzerland": "Suiza",
    "England": "Inglaterra",
    "Spain": "España",
    "United States": "USA",
    "Korea Republic, Japan": "Corea del Sur, Japón",
    "South Africa": "Sudáfrica",
    "Russia": "Rusia",
    "Qatar": "Catar"
}
```

```python
df_mundiales[["Champion", "Runner-Up", "Host"]] = df_mundiales[["Champion", "Runner-Up", "Host"]].replace(paises)
df_mundiales
```

## Selección de las columnas para nuestro nuevo *Dataframe*

Para ello, también usaremos la función *.copy()* para crear uno independiente del original.

```python
columnas_original_permanecen = ["Year", "Host", "Champion", "Runner-Up", "TopScorrer"]
df_nuevo = df_mundiales[columnas_original_permanecen].copy()
```

## Agrupación de valores en función de una columna

Se agrupan los valores en cada columna en función del campeón para poder establecer el récord de cada selección campeona, los años que ganó, el subcampeón, el máximo goleador de la edición y el anfitrión donde se proclamó campeona dicha selección.

```python
df_nuevo = df_nuevo.groupby("Champion", as_index= False).agg(
    winner = ("Champion", "count"),
    year = ("Year", lambda year: ", ".join(year.astype(str))),
    runner_up = ("Runner-Up", ", ".join),
    top_scorer = ("TopScorrer", ", ".join),
    host = ("Host", ", ".join)
)


df_nuevo

```

## Renombrar columnas a español

```python
columnas_nuevas = {
    "Champion": "Campeón",
    "year": "Años",
    "host": "Anfitriones",
    "runner_up": "Subcampeones",
    "top_scorer": "Máximo goleador",
    "winner": "Copas"
}

df_nuevo = df_nuevo.rename(columns=columnas_nuevas)
df_nuevo
```

## Exportar el *DataFrame* generado como un *archivo.csv*

```python
df_nuevo.to_csv("Estadísticas_mundiales.csv", index=False, encoding="utf-8-sig")


```
